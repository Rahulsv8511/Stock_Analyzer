```python
import matplotlib.pyplot as plt
import numpy as np
# Sample data

np.random.seed(0)

# Data for histogram (e.g., ages of individuals)

ages = np.random.normal(loc=30, scale=5, size=500)  # 500 ages around 30 years old

# Data for scatter plot (e.g., size vs price)

house_size = np.random.normal(loc=1500, scale=300, size=100)  # sizes in square feet

price = house_size * 200 + np.random.normal(0, 20000, size=100)  # price = 200 * size + noise

# 1. Histogram of ages

plt.figure(figsize=(4,3))

plt.hist(ages, bins=20, color='skyblue', edgecolor='black')

plt.title("Histogram of Ages")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()

# 2. Box plot of prices in two fictional neighborhoods

prices_neighborhoodA = 200000 + np.random.normal(0, 30000, size=100)

prices_neighborhoodB = 180000 + np.random.normal(0, 25000, size=100)

plt.figure(figsize=(4,3))

plt.boxplot([prices_neighborhoodA, prices_neighborhoodB], labels=["Neighborhood A", "Neighborhood B"])

plt.ylabel("House Price")

plt.title("Price Distribution by Neighborhood")

plt.show()

# 3. Scatter plot of house size vs price

plt.figure(figsize=(4,3))

plt.scatter(house_size, price, color='purple', alpha=0.7)

plt.title("Scatter Plot: House Size vs Price")

plt.xlabel("Size (sq ft)")

plt.ylabel("Price ($)")

plt.show()
```

```python
import numpy as np

# Discrete distribution example: Loaded die (probabilities for faces 1-6)

# Let's say the die is biased such that 6 is twice as likely as each other face.

# Probabilities: 1-5 have p=1/7 each, 6 has p=2/7 (sums to 1).

probabilities = [1/7]*5 + [2/7]  # for faces 1,2,3,4,5,6

faces = np.arange(1, 7)

# Draw 100000 samples from this distribution

samples = np.random.choice(faces, size=100000, p=probabilities)

# Empirical probability for each face:

for face in faces:

    emp_prob = np.mean(samples == face)

    print(f"Empirical P(face={face}) = {emp_prob:.3f}")

print("---")

# Continuous distribution example: Normal distribution with mean=0, std=1

cont_samples = np.random.normal(loc=0.0, scale=1.0, size=100000)

print(f"Sample mean ~ {np.mean(cont_samples):.3f}, sample std ~ {np.std(cont_samples):.3f}")

# Let's check roughly the fraction of samples within 1 and 2 standard deviations

within_1_std = np.mean((cont_samples >= -1) & (cont_samples <= 1))

within_2_std = np.mean((cont_samples >= -2) & (cont_samples <= 2))

print(f"Fraction of samples within 1 std of mean: {within_1_std:.3f}")

print(f"Fraction of samples within 2 std of mean: {within_2_std:.3f}")
```

```python
import numpy as np

# Define a "population"

population = np.arange(1, 10001)  # population = numbers 1 to 10000 (population size N=10000)

pop_mean = np.mean(population)

pop_std = np.std(population)

print(f"Population mean = {pop_mean:.2f}, Population std = {pop_std:.2f}")

# Take a random sample of 100 from the population

sample = np.random.choice(population, size=100, replace=False)

sample_mean = np.mean(sample)

sample_std = np.std(sample, ddof=1)  # use ddof=1 for sample std (unbiased estimator)

print(f"Sample mean = {sample_mean:.2f}, Sample std = {sample_std:.2f}")

# Take a larger sample of 1000

sample2 = np.random.choice(population, size=1000, replace=False)

print(f"Larger sample mean = {np.mean(sample2):.2f}, std = {np.std(sample2, ddof=1):.2f}")

```

```python
import numpy as np

# Original distribution: exponential (which is skewed, not normal)

lam = 1.0  # parameter for exponential (mean = 1/lam = 1.0)

true_mean = 1/lam

print(f"True mean of exponential distribution = {true_mean:.2f}")

def sample_means(dist_func, sample_size, n_means):

    means = []

    for _ in range(n_means):

        data = dist_func(size=sample_size)

        means.append(np.mean(data))

    return np.array(means)

# Generate distribution of sample means for different sample sizes

for n in [1, 5, 20, 50]:

    means = sample_means(lambda size: np.random.exponential(1/lam, size=size), sample_size=n, n_means=10000)

    print(f"Sample size {n}: mean of means = {means.mean():.3f}, std of means = {means.std():.3f}")
```
